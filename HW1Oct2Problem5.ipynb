```julia
using Convex
using ECOS
using PyPlot
```

```julia

m=30;
n=100;
A=randn(m,n);
b=randn(m,1);
```

```julia
x1 = Variable(n);
p = minimize(sum_squares(A*x1-b));
solve!(p,ECOSSolver())
```

```julia
figure(figsize=(20,5));
r1 = -1:1/50:1;
plt[:hist](x1.value,50,facecolor="w")
plot(r1,r1.^2,"k")
axis([-1,1,0,10]);
```

```julia
x2 = Variable(n)
p = minimize(sum(abs(A*x2-b)))
solve!(p,ECOSSolver())
```

```julia
figure(figsize=(20,5));
r1 = -1:0.1/50:1;
plt[:hist](x2.value,50,facecolor="w")
plot(r1,abs(r1),"k")
axis([-1,1,0,10]);
```

```julia
x3 = Variable(n)
p = minimize(sum(max(abs(A*x3-b) - .05*ones(b),zeros(b))))
solve!(p,ECOSSolver())
```

```julia
figure(figsize=(20,5));
r1 = -1:0.1/50:1;
plt[:hist](x3.value,50,facecolor="w")
plot(r1,max(abs(r1)-.5,0),"k")
axis([-1,1,0,10]);
```

```julia
x4 = Variable(n)
p = minimize(sum(-log(1-x4)- log(1+x4)))
solve!(p,ECOSSolver())
```

```julia
figure(figsize=(20,5));
r1 = -1:0.1/50:1;
plt[:hist](x4.value,50,facecolor="w")
plot(r1,-log(1+r1) - log(1-r1),"k")
axis([-1,1,0,10]);
```
